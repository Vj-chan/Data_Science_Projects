### Pre-processing:

**Handling Missing Values**

Lets find the number of missing values in all the columns,

```python
df.isna().sum()
```

The duration column has only 3 numbers of missing values. Let's first take a look at those rows and try to fill those values,

```python
df[df['duration'].isna()]
```

It looks like the duration for those rows are mistakenly entered in rating column. Let's fix that,

```python
df['duration'].fillna((df.loc[df['duration'].isna(),'rating']), inplace = True)
```

Now, lets check for the changes and also change the values in rating column

```python
df.iloc[[5541,5794,5813]]
```

```python
df.loc[[5541,5794,5813],'rating'] = 'unknown'
df.loc[[5541,5794,5813]]
```

Next, lets fill out the missing values in date_added column. Before filling lets take a look at the rows where the date_added values are missing, 

```python
df.loc[df['date_added'].isna()]
```

To fill the date_added column, we cannot fill it with random date but also we don't have accurate date. So, to make it more relevant, lets find the date in which most of the tv shows added for all released year,

```python
dateref = pd.DataFrame(df.loc[df['type'] == 'TV Show'].groupby('release_year')['date_added'].value_counts())
```

```python
dateref = dateref.rename(columns={'date_added':'date_count'}).reset_index()
```

```python
dateref.sort_values(by='date_count',ascending = False, inplace = True)
```

```python
dateref.drop_duplicates(subset = 'release_year', inplace = True)
```

```python
dateref.reset_index(drop = True, inplace = True)
```

```python
dateref
```

Now we have a reference table consist of the date for every year in which most of the tv shows are added. Lets fill them in our main dataframe.

To fill them, first we need to merge the reference table with the main dataframe and extract the date added values from the reference table for the missing values in the main dataframe and store them in a separate variable.

```python
datefill = df.merge(dateref, 
                    left_on = 'release_year', 
                    right_on = 'release_year', 
                    how = 'left', 
                    suffixes = [None, "_dr"]).loc[df['date_added'].isna()]['date_added_dr']
```

```python
datefill
```

```python
df['date_added'].fillna(datefill, inplace = True)
```

```python
df.isna().sum()
```

The rating column also has some null values. As we have already filled some values as unknown, so lets fill these also with the same,

```python
df['rating'].fillna('unknown', inplace = True)
```

```python
df.isna().sum()
```

Let's fill the missing values in the columns - director, cast, country with 'unknown'. Because it does not make any sense to fill those values with any values other than exact values

```python
df['director'].fillna('unknown', inplace = True)
df['cast'].fillna('unknown', inplace = True)
df['country'].fillna('unknown', inplace = True)
```

Lets confirm whether all the missing values are filled,

```python
df.isna().sum()
```

Let's convert the data types of the columns into appropriate data type :

The data type of the column 'date_added' is in 'object' dtype. So lets convert it into datetime format

```python
df.loc[:,['date_added']] = pd.to_datetime(df['date_added'])
```

Let's check whether the datatype has changed or not

```python
df.info()
```

Now the datatype of the 'date_added' column has changed to 'datetime64' format

The 'duration' column has units along with the numbers and also in string format. Lets extract the numbers and convert them into float dtype.

```python
df['duration'] = df['duration'].apply(lambda x : float(str(x).split(' ')[0]))
```

Let's check whether the number is properly extracted

```python
df.head()
```

The columns 'director', 'cast', 'country', 'listed_in' has nested values in single string. Lets split and add them in separate rows.

Splitting the nested values into list

```python
df['director'] = df['director'].str.split(', ')
df['cast'] = df['cast'].str.split(', ')
df['country'] = df['country'].str.split(', ')
df['listed_in'] = df['listed_in'].str.split(', ')
```

Adding list of values into separate rows

```python
df = df.explode('director', ignore_index = True)
df = df.explode('cast', ignore_index = True)
df = df.explode('country', ignore_index = True)
df = df.explode('listed_in', ignore_index = True)
```

```python
df.head()
```

Lets check for any duplicate values,

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep = 'first', inplace = True)
```

```python
df.duplicated().sum()
```

Renaming the column 'listed_in' to 'genre'

```python
df.rename(columns={'listed_in' : 'genre'}, inplace = True)
```

Creating a new dataframe with only necessary columns in new order

```python
df_cleaned = df.loc[:,['show_id', 'type', 'title', 'genre', 'release_year', 'date_added', 
                       'director', 'cast', 'country', 'rating', 'duration']]
```

```python
df_cleaned.head()
```

Now we have clean data to proceed the analysis.
